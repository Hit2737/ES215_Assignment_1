```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('data1.txt', sep=' ', header=None)
```

```python
data.columns = ['(a) Rec', '(b) Loop', '(c) Rec+mem', '(d) Loop+mem']
data.index = data.index + 1
```

```python
data
```

```python
c = ['r--o', 'g--o', 'b--o', 'y--o']
plt.subplots(figsize=(10, 7))
plt.suptitle('Time vs Iteration')
for i in range(4):
    plt.subplot(2,2,i+1)
    plt.plot(data[data.columns[i]], c[i]);
    plt.title(data.columns[i])
    plt.grid()
    plt.xticks(range(1, 6, 1))
    plt.xlabel('Iteration')
    plt.ylabel('Time (s)')
plt.tight_layout()
plt.show()
```

```python
print("Average time for each method:")
data.mean()
```
